```python
import sys
sys.path.append('../')
import bandit #importa o ambient
import numpy as np
```

## Definindo a função argmax

```python
def argmax(Q_values):
    """ (lista) -> índice de maior valor(int)
    Recebe uma lista dos calores Q e retorna o índice do maior valor. Por definição resolve empates escolhendo um deles aleatóriamente.
    """
    max_value = float("-inf")
    ties = []

    for i in range(len(Q_values)):
        if Q_values[i] > max_value:
            max_value = Q_values[i]
            ties = []
        if Q_values[i] == max_value:
            ties.append(i)

    return np.random.choice(ties) 
```

## Definindo o Agente

```python
class Greedy(object):
    def __init__(self, k_arms):
        self.n_arms = np.zeros(k_arms)
        self.Q_values = np.zeros(k_arms)
        self.last_action = argmax(self.Q_values)

    def agent_step(self, reward):
        """ (float) -> acao(int)
        Dá um step para o Agente atualizando os valores Q. Pega a recompensas do estado e retorna a ação escolhida.
        """ 

        self.n_arms[self.last_action] += 1
        step_size = 1/self.n_arms[self.last_action]
        current_Q = self.Q_values[self.last_action] + step_size * (reward - self.Q_values[self.last_action])
        self.Q_values[self.last_action] = current_Q

        current_action = argmax(self.Q_values)

        return current_action
```

## Testando o Agente

```python
import matplotlib.pyplot as plt
```

```python
num_runs = 200
num_steps = 1000
np.random.seed(1)
total_means = []

max_means = 0
for runs in range(num_runs):

    agent = Greedy(k_arms=10)

    env = bandit.GaussianBandit(k_arms=10)

    score = [0]
    means = []

    max_means += np.max(env.bandits_expectations)

    for steps in range(num_steps):
        rewards = env.gamble(agent.last_action)
        agent.last_action = agent.agent_step(rewards)

        score.append(score[-1] + rewards)
        means.append(score[-1]/(steps+1))
    total_means.append(means)

plt.figure(figsize=(15,5))
plt.plot([max_means / num_runs for _ in range(num_steps)], linestyle="--")
plt.plot(np.mean(total_means, axis=0))
plt.legend(["Melhor Possível", "Guloso"])
plt.title("Recompensa Média de um agent Guloso")
plt.xlabel("Steps")
plt.ylabel("Recompensa Média")
plt.show()
```

Aqui fica claro a ineficiência do algoritmo, mostrando como **exploração** é importante.
